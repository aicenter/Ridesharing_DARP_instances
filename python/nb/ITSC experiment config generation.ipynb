```python
import os
import numpy as np
import math
from pathlib import Path
from typing import List
from datetime import timedelta, datetime

import darpinstances.experiments
import darpinstances.instance
```

```python
darp_dir = Path(r'C:\Google Drive\AIC Experiment Data\DARP\ITSC_instance_paper')
```

```python
instance_dir = darp_dir / 'Instances'
result_dir = darp_dir / 'Results'
time_limit_multiplier = 1
time_buffer_for_dm = 300
methods = {
    "ih": {"method": "ih"},
    'halns': {"method": "halns"},
    'halns-ih': {"method": "halns", 'init': '../ih/config.yaml-solution.json'},
    "vga": {"method": "vga", 'tmax': 60}
}
```

```python
for root, dir, files in os.walk(instance_dir):
    for file in files:
        filename = os.fsdecode(file)
        if filename == "config.yaml":
            filepath = Path(root) / filename

            index = filepath.parts.index('Instances')
            result_path = result_dir / Path(*filepath.parts[index + 1:index + 2]) / Path(*filepath.parts[index + 3:-1])

            # time limit configuration
            config = darpinstances.instance.load_instance_config(filepath)
            duration = datetime.strptime(config['demand']['max_time'], '%Y-%m-%d %H:%M:%S') \
                       - datetime.strptime(config['demand']['min_time'], '%Y-%m-%d %H:%M:%S')
            duration_seconds = duration.total_seconds()
            for _, method in methods.items():
                method['timeout'] = duration_seconds * time_limit_multiplier + time_buffer_for_dm

            darpinstances.experiments.generate_experiment_configs_for_instance(result_path, methods, filepath, False)
```
